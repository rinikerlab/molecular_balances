# Molecular balances: MM vs. QM/MM
Analysis of results

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
# plot free energies
fig, ax = plt.subplots(1, 3, sharex=True, sharey=True, figsize=(45,15), gridspec_kw={'hspace': 0.2, 'wspace': 0.2})
fig.gca().xaxis.set_major_locator(plt.MultipleLocator(60))
plt.minorticks_on()

# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_mm_balance1.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[0].set_title(f"Balance 1", fontsize=50)

ax[0].plot(angle, free_energy, alpha=0.7, color='green', linewidth=5, label='MM')
ax[0].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='green', linewidth=3)
ax[0].grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)

# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_qmmm_balance1.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[0].plot(angle, free_energy, alpha=0.7, color='red', linewidth=5, label='QM/MM')
ax[0].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='red', linewidth=3)


# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_mm_balance4.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[1].set_title(f"Balance 4", fontsize=50)
ax[1].plot(angle, free_energy, alpha=0.7, color='green', linewidth=5, label='MM')
ax[1].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='green', linewidth=3)
ax[1].grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)


# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_qmmm_balance4.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[1].plot(angle, free_energy, alpha=0.7, color='red', linewidth=5, label='QM/MM')
ax[1].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='red', linewidth=3)
ax[1].legend(fontsize=30, loc='upper center', frameon=False)


# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_mm_balance6.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[2].set_title(f"Balance 6", fontsize=50)
ax[2].plot(angle, free_energy, alpha=0.7, color='green', linewidth=5, label='MM')
ax[2].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='green', linewidth=3)
ax[2].grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)

# get free energy from WHAM
angle = []
free_energy = []
error = []
with open (f"freeene_qmmm_balance6.txt", "r") as f:
    for line in f:
        if line.split()[0][0] !='#':
            angle.append(float(line.split()[0]))
            free_energy.append(float(line.split()[1]))
            error.append(float(line.split()[2]))

free_energy = np.array(free_energy) - min(free_energy)
ax[2].plot(angle, free_energy, alpha=0.7, color='red', linewidth=5, label='QM/MM')
ax[2].fill_between(angle, free_energy - error, free_energy + error, alpha=0.3, color='red', linewidth=3)

fig.add_subplot(111, frame_on=False)
plt.xlabel("Pseudo-dihedral angle []", fontsize=50)
plt.ylabel("Potential of mean force [kJ/mol]", fontsize=50, labelpad=30)
plt.tick_params(labelcolor='None', bottom=False, left=False, labelsize=30)
for axes in ax:
    axes.tick_params(axis='both', which='major', labelsize=30)

ax[2].set_xlim(-180,180)
ax[2].set_ylim(0,22)
#plt.savefig("pmf_comparison.pdf", bbox_inches='tight', transparent=True)
```

```python
# values from WHAM
data = np.load("wham_free_energies.npy", allow_pickle=True).item()

dG_mm1 = data["dG_mm1"]
dG_mm1_error = data["dG_mm1_error"]

dG_mm2 = data["dG_mm2"]
dG_mm2_error = data["dG_mm2_error"]

dG_qmmm1 = data["dG_qmmm1"]
dG_qmmm1_error = data["dG_qmmm1_error"]

dG_qmmm2 = data["dG_qmmm2"]
dG_qmmm2_error = data["dG_qmmm2_error"]

dG_qmmm1gfn1 = data["dG_qmmm1gfn1"]
dG_qmmm1gfn1_error = data["dG_qmmm1gfn1_error"]

dG_qmmm2gfn1 = data["dG_qmmm2gfn1"]
dG_qmmm2gfn1_error = data["dG_qmmm2gfn1_error"]

dG_qmmm1dftb = data["dG_qmmm1dftb"]
dG_qmmm1dftb_error = data["dG_qmmm1dftb_error"]

dG_qmmm2dftb = data["dG_qmmm2dftb"]
dG_qmmm2dftb_error = data["dG_qmmm2dftb_error"]
```

```python
dG_mm = (dG_mm1 + dG_mm2)/2
dG_mm_err = np.sqrt((dG_mm1_error**2 + dG_mm2_error)/4)

dG_qmmm = (dG_qmmm1 + dG_qmmm2)/2
dG_qmmm_err = np.sqrt((dG_qmmm1_error**2 + dG_qmmm2_error)/4)

dG_qmmmgfn1 = (dG_qmmm1gfn1 + dG_qmmm2gfn1)/2
dG_qmmmgfn1_err = np.sqrt((dG_qmmm1gfn1_error**2 + dG_qmmm2gfn1_error)/4)

dG_qmmmdftb = (dG_qmmm1dftb + dG_qmmm2dftb)/2
dG_qmmmdftb_err = np.sqrt((dG_qmmm1dftb_error**2 + dG_qmmm2dftb_error)/4)
```

```python
#######
# exp #
#######
experimental50 = np.array([0.88, -0.71, -0.13, 0.00, 0.71, 1.55, 1.84])
experimental100 = np.array([0.92, -0.67, -0.42, 0.00, 1.00, 1.88, 2.47])

experimental = (experimental50 + experimental100) / 2

# MM US
spearman_mm = sc.stats.spearmanr(experimental, dG_mm).statistic
kendall_mm = sc.stats.kendalltau(experimental, dG_mm).statistic
pearson_mm, p_value = sc.stats.pearsonr(experimental, dG_mm)
rmse_mm = np.sqrt(np.mean((dG_mm-experimental)**2))
mae_mm = np.mean(np.abs(experimental - dG_mm))
print(f"MM US: Spearman - {spearman_mm}, Kendall - {kendall_mm}, Pearson - {pearson_mm}, RMSE - {rmse_mm}, MAE - {mae_mm}")

# QM/MM GFN1-xTB US
spearman_qmmm_gfn1 = sc.stats.spearmanr(experimental, dG_qmmmgfn1).statistic
kendall_qmmm_gfn1 = sc.stats.kendalltau(experimental, dG_qmmmgfn1).statistic
pearson_qmmm_gfn1, p_value = sc.stats.pearsonr(experimental, dG_qmmmgfn1)
rmse_qmmm_gfn1 = np.sqrt(np.mean((dG_qmmmgfn1-experimental)**2))
mae_qmmm_gfn1 = np.mean(np.abs(experimental - dG_qmmmgfn1))
print(f"QM/MM GFN1-xTB US: Spearman - {spearman_qmmm_gfn1}, Kendall - {kendall_qmmm_gfn1}, Pearson - {pearson_qmmm_gfn1}, RMSE - {rmse_qmmm_gfn1}, MAE - {mae_qmmm_gfn1}")

# QM/MM GFN2-xTB US
spearman_qmmm = sc.stats.spearmanr(experimental, dG_qmmm).statistic
kendall_qmmm = sc.stats.kendalltau(experimental, dG_qmmm).statistic
pearson_qmmm, p_value = sc.stats.pearsonr(experimental, dG_qmmm)
rmse_qmmm = np.sqrt(np.mean((dG_qmmm-experimental)**2))
mae_qmmm = np.mean(np.abs(experimental - dG_qmmm))
print(f"QM/MM GFN2-xTB US: Spearman - {spearman_qmmm}, Kendall - {kendall_qmmm}, Pearson - {pearson_qmmm}, RMSE - {rmse_qmmm}, MAE - {mae_qmmm}")

# QM/MM DFTB3 US
spearman_qmmm_dftb = sc.stats.spearmanr(experimental, dG_qmmmdftb).statistic
kendall_qmmm_dftb = sc.stats.kendalltau(experimental, dG_qmmmdftb).statistic
pearson_qmmm_dftb, p_value = sc.stats.pearsonr(experimental, dG_qmmmdftb)
rmse_qmmm_dftb = np.sqrt(np.mean((dG_qmmmdftb-experimental)**2))
mae_qmmm_dftb = np.mean(np.abs(experimental - dG_qmmmdftb))
print(f"QM/MM DFTB3 US: Spearman - {spearman_qmmm_dftb}, Kendall - {kendall_qmmm_dftb}, Pearson - {pearson_qmmm_dftb}, RMSE - {rmse_qmmm_dftb}, MAE - {mae_qmmm_dftb}")

plt.figure(figsize=(12,12))
#plt.title("Correlation between experimental and REST2/OpenFF-calculated free energy differences", fontsize=18)
plt.xlim(-4.5,6.5)
plt.ylim(-4.5,6.5)

# plot
plt.plot(experimental, dG_mm, linestyle='none', marker='x', markeredgewidth=5, markersize=15, color='green', alpha=0.8, label='MM (OpenFF)')
plt.plot(experimental, dG_qmmmgfn1, linestyle='none', marker='o', markersize=15, color='orange', alpha=0.8, label='QM/MM (GFN1-xTB)')
plt.plot(experimental, dG_qmmm, linestyle='none', marker='o', markersize=15, markerfacecolor='None', markeredgecolor='red', markeredgewidth=4, alpha=0.8, label='QM/MM (GFN2-xTB)')
plt.plot(experimental, dG_qmmmdftb, linestyle='none', marker='s', markersize=15, color='blue', alpha=0.8, label='QM/MM (DFTB3)')

# annotate
plt.text(experimental[0], 2.0, '1', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[1], -3.85, '2', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[2], -3.65, '3', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[3], -0.9, '4', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[4], -0.1, '5', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[5], 1.9, '6', fontsize=20, horizontalalignment='center', verticalalignment='center')
plt.text(experimental[6], 3.2, '7', fontsize=20, horizontalalignment='center', verticalalignment='center')

plt.plot([-4.5,6.5], [-4.5,6.5], color='black', alpha=0.7)
plt.plot([-4.5,6.5], [-4.5,6.5], color='grey', alpha=0.15, lw=515)
plt.xlabel("G$^{experimental}$ [kJ/mol]", fontsize=35)
plt.ylabel("G$^{calculated}$ [kJ/mol]", fontsize=35)
plt.tick_params(labelsize=30)
plt.legend(fontsize=20, frameon=False)

#plt.savefig("correlation.pdf", dpi=600, bbox_inches='tight', transparent=True)
```
